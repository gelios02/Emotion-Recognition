Распознавания эмоций человека для психоанализа с помощью методы глубокого обучения и компьютерного зрения

Здесь мы импортируем необходимые библиотеки: numpy для работы с массивами данных, pandas для обработки данных в формате CSV, tensorflow для создания и обучения модели глубокого обучения, а также несколько модулей из tensorflow.keras для создания архитектуры модели.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```



1. Импорт необходимых модулей:
   - `pandas` (под именем `pd`) для работы с данными в формате CSV.
   - `numpy` (под именем `np`) для работы с массивами и матрицами.
   - `tensorflow` (под именем `tf`) для создания и обучения модели глубокого обучения.
   - `train_test_split` из модуля `sklearn.model_selection` для разделения данных на обучающую и тестовую выборки.

2. Загрузка данных из файла CSV:
   - Используется функция `read_csv` из модуля `pandas` для чтения данных из файла 'sample_data/fer2013.csv'.
   - Пиксели изображений сохраняются в переменную `pixels` в виде списка.
   - Метки эмоций сохраняются в переменную `labels` в виде списка.

3. Преобразование пикселей и меток:
   - Пиксели преобразуются в массивы с помощью функции `split()` и преобразования типов данных.
   - Пиксели нормализуются, чтобы значения находились в диапазоне от 0 до 1.
   - Пиксели переформатируются в трехмерный массив размерности (количество изображений, ширина, высота, каналы).
   - Метки эмоций преобразуются в формат one-hot с помощью функции `to_categorical` из модуля `tf.keras.utils`.

4. Разделение данных на обучающую и тестовую выборки:
   - Используется функция `train_test_split` для разделения пикселей и меток на обучающую и тестовую выборки.
   - Параметр `test_size=0.2` указывает, что 20% данных будет использоваться для тестирования.

5. Создание модели глубокого обучения:
   - Используется `Sequential` модель из `tensorflow.keras` для создания последовательной нейронной сети.
   - Добавляются слои свертки (`Conv2D`) с функцией активации `relu` и пуллинга (`MaxPooling2D`) для извлечения признаков.
   - Добавляются слои плоского представления (`Flatten`) и полносвязные слои (`Dense`) с функцией активации `relu` и `softmax`.
   - Используется 7 выходных нейронов для классификации 7 эмоций.

6. Компиляция модели:
   - Указывается оптимизатор `adam`, функция потерь `categorical_crossentropy` и метрика `accuracy`.

7. Обучение модели:
   - Используется метод `fit` для обучения модели на обучающих данных.
   - Указывается количество эпох, размер пакета (batch_size) и данные для валидации.

8. Сохранение модели:
   - Используется метод `save` для сохранения обученной модели в формате .h5.

Этот код выполняет обучение модели для распознавания эмоций на датасете fer2013.csv и сохраняет полученную модель в файл 'emotion_model.h5'.

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Загрузка данных из файла CSV
data = pd.read_csv('sample_data/fer2013.csv')

# Извлечение пикселей и меток
pixels = data['pixels'].tolist()
labels = data['emotion'].tolist()

# Преобразование пикселей в массивы
pixels = [np.array(p.split(), dtype='float32') for p in pixels]
pixels = np.array(pixels) / 255.0
pixels = pixels.reshape(pixels.shape[0], 48, 48, 1)

# Преобразование меток в формат one-hot
labels = tf.keras.utils.to_categorical(labels, num_classes=7)

# Разделение данных на обучающую и тестовую выборки
train_images, test_images, train_labels, test_labels = train_test_split(pixels, labels, test_size=0.2, random_state=42)

# Создание модели
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(7, activation='softmax')
])

# Компиляция модели
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Обучение модели
model.fit(train_images, train_labels, epochs=10, batch_size=64, validation_data=(test_images, test_labels))

model.save('emotion_model.h5')



```

Данный код выполняет следующие действия:

1. Импорт необходимых модулей:
   - `numpy` (под именем `np`) для работы с массивами и матрицами.
   - `cv2` для работы с изображениями и обработки изображений.
   - `matplotlib.pyplot` (под именем `plt`) для визуализации изображений.
   - `tensorflow.keras.models.load_model` для загрузки сохраненной модели.

2. Загрузка сохраненной модели:
   - Используется функция `load_model` для загрузки модели из файла 'emotion_model.h5'.

3. Запрос пути к изображению через ввод пользователем:
   - С помощью функции `input` запрашивается путь к изображению.

4. Загрузка и предобработка изображения:
   - Изображение загружается с помощью функции `cv2.imread` и сохраняется в переменную `image`.
   - Используется функция `cv2.cvtColor` для преобразования цветовой схемы изображения в RGB.
   - Изображение отображается с помощью `plt.imshow` и `plt.show`.
   - Изображение преобразуется в оттенки серого с помощью `cv2.cvtColor`.
   - Изображение изменяется размером до 48x48 пикселей с помощью `cv2.resize`.
   - Изображение расширяется в одну размерность с помощью `np.expand_dims`.
   - Изображение нормализуется, чтобы значения пикселей находились в диапазоне от 0 до 1.

5. Предсказание эмоции:
   - Модель используется для предсказания эмоции на загруженном изображении с помощью `model.predict`.
   - Полученные предсказания сохраняются в переменную `predictions`.
   - Используется функция `np.argmax` для определения индекса с максимальным значением в предсказаниях, соответствующему эмоции.
   - Полученный индекс соответствует метке эмоции.

6. Сопоставление метки эмоции с текстовым описанием:
   - Создается словарь `emotion_mapping`, где ключами являются числовые метки эмоций, а значениями - текстовые описания эмоций.
   - Метка эмоции полученная из предсказания сопоставляется с текстовым описанием эмоции с помощью словаря `emotion_mapping`.

7. Отображение изображения с подписью эмоции:
   - Изображение отображается с помощью `plt.imshow`.
   - Картинка представляется в оттенках серого с помощью `cv2.convertScaleAbs`.
   - В заголовок графика добавляется текстовое описание эмоции с помощью `plt.title`.
   - Оси координат выключаются с помощью `plt.axis('off')`.
   - Изображение с эмоцией отображается с помощью `plt.show`.

8. Вывод определенной эмоции:
   - С помощью `print` выводится определенная эмоция.

Этот код загружает предварительно обученную модель распознавания эмоций, запрашивает путь к изображению от пользователя, загружает и обрабатывает изображение, предсказывает эмоцию на изображении, выводит изображение с указанной эмоцией и выводит текстовое описание определенной эмоции.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Загрузка сохраненной модели
model = load_model('emotion_model.h5')

# Запрос изображения через input
image_path = input("Введите путь к изображению: ")

# Загрузка и предобработка изображения
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = cv2.resize(image, (48, 48))
image = np.expand_dims(image, axis=-1)
image = image / 255.0

# Предсказание эмоции
predictions = model.predict(np.array([image]))
emotion_label = np.argmax(predictions[0])

# Сопоставление метки эмоции с текстовым описанием
emotion_mapping = {
    0: "Angry",
    1: "Disgust",
    2: "Fear",
    3: "Happy",
    4: "Sad",
    5: "Surprise",
    6: "Neutral"
}

emotion = emotion_mapping[emotion_label]
image_rgb = cv2.convertScaleAbs(image)
plt.imshow(image_rgb, cmap='gray')
plt.title(emotion)
plt.axis('off')
plt.show()
print("Определенная эмоция: ", emotion)
```
